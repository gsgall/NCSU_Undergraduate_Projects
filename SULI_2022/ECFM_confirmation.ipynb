```python
import numpy as np 
import random as rand 
from numpy import genfromtxt
import matplotlib.pyplot as plt 
```

```python
matrix_file = 'ECFM_matrix.csv'
frequency_file = 'ECFM_freq.csv'
position_file = 'ECFM_rad.csv'

A = np.transpose(np.vectorize(lambda t:complex(t.replace('i','j'))) (genfromtxt(matrix_file, delimiter=',',dtype='str')))
v =  v = [1.3, 1.2, 1.15, 1.1, 1.05, 1, .95] # m / s 
rad_pos = np.loadtxt(position_file, delimiter=',') # m
freq = np.loadtxt(frequency_file, delimiter=',') # Hz 

num_sens = A.shape[0]
num_vel = A.shape[1]

sensitivity = np.abs(np.matmul(A, v))
plt.figure() 
plt.plot(freq, sensitivity)
plt.title('Sensitivity versus frequency')
plt.ylabel('Sensitivity (V)')
plt.xlabel('Drive Frequency Hz')
plt.show() 
```
