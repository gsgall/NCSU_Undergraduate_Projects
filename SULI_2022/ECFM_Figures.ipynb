```python
import matplotlib.pyplot as plt 
import numpy as np 
# Updating Default Plot Font to be LaTex
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
val = 1
width = val * 0.05
num_samples = 1000 
input = np.random.normal(val, width, num_samples).reshape(-1, 1)

mu, sigma = 0, 0.05 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 30, density=True, edgecolor='black', range=(-0.16, 0.16))
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.ylabel('Counts')
plt.xlabel('Noisy Sensitivity Value')
plt.title('Example of Noise Distribution')
plt.savefig('example.pdf', format='pdf', bbox_inches='tight')
plt.show()

plt.plot(s)
print(1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ))
```
